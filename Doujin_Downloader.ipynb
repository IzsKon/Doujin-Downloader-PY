```python
import requests
from bs4 import BeautifulSoup
import os
import shutil

def parse_soup(url):
    html = requests.get(url).text
    return BeautifulSoup(html, 'html.parser')

# Get main doujin page
magic_num = input('Magic Number: ')
url = f'https://nhentai.net/g/{magic_num}'
html = parse_soup(url)

# Get title
title = html.find_all('span', class_='pretty')[-1].text
print(f'Title: {title}')

# Get numnber of pages
html = parse_soup(f'{url}/1')
num_pages = int( html.find_all(class_='num-pages')[0].text )
print(f'Pages: {num_pages}')

# Create directory for download
dir = f'{title}/{title}/'
os.makedirs(os.path.dirname(dir), exist_ok=True)

# Start download
print('\n[ Start Download ]')

for page in range(1, num_pages+1):
    print(f'Donwloading page {page} ...')
    html = parse_soup(f'{url}/{page}')
    src = html.select('#image-container img')[0].get('src')
    ext = os.path.splitext(src)[1]
    img = requests.get(src, allow_redirects=True)
    with open(f'{dir}{page}{ext}', 'wb') as f:
        f.write(img.content)

print('[ Download Complete ]')
shutil.make_archive(title, 'zip', title)
```
